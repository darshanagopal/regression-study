```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 
```

##### pandas is used for data manipulation and analysis.
##### LinearRegression from sklearn is used to create a linear regression model.
##### matplotlib.pyplot is used for plotting data.
##### train_test_split from sklearn is used to split the data into training and testing sets.


##### Load the dataset into a pandas DataFrame and display the first 3 rows to inspect the data structure.


```python
#import data
data = pd.read_csv('/Users/darshanagopal/Desktop/Code/Regression/data/Advertising.csv')
print(data.head(3))
```


##### Split the data into training (80%) and testing (20%) sets using train_test_split.
##### Set random_state for reproducibility.

```python
# Split dataset into training and testing sets
train, test = train_test_split(data, test_size=0.2, random_state=42)
print('Train set:\n', train.head(3))  # Display the first 3 rows of the training set
print('Test set:\n', test.head(3))    # Display the first 3 rows of the testing set
```


##### Separate the features (independent variables) and the target variable (dependent variable) for both training and testing sets.

```python
# Define features (independent variables) and target variable (dependent variable)
X_train = train[['TV', 'Radio', 'Newspaper']]  # Features for training
Y_train = train['Sales']                       # Target variable for training
X_test = test[['TV', 'Radio', 'Newspaper']]    # Features for testing
Y_test = test['Sales']                         # Target variable for testing
```



##### Create an instance of the LinearRegression model and fit it to the training data.

```python
# Create a linear regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(X_train, Y_train)

# Predict the target variable (Sales) on the test set
sales_pred = model.predict(X_test)
print('Predicted Sales:\n', sales_pred)  # Print the predicted sales values
```

##### Create a scatter plot comparing actual sales (Y_test) to predicted sales (sales_pred).
##### Plot a red dashed line representing perfect prediction (where actual equals predicted).
##### Add labels, title, and legend to the plot, and display it.

```python
# Plotting actual vs predicted sales
plt.figure(figsize=(10, 6))
plt.scatter(Y_test, sales_pred, alpha=0.7, color='blue', label='Predicted vs Actual')  # Scatter plot of actual vs predicted sales
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], '--', color='red', label='Ideal Fit')  # Line representing perfect prediction
plt.xlabel('Actual Sales')  # Label for x-axis
plt.ylabel('Predicted Sales')  # Label for y-axis
plt.title('Actual Sales vs Predicted Sales')  # Title of the plot
plt.legend()  # Show legend
plt.show()  # Display the plot
```
