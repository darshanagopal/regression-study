```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('/Users/darshanagopal/Library/Mobile Documents/com~apple~CloudDocs/Code/Regression/data/Fish.csv')
```

```python
print(data.head(4))
```

```python
y = data['Weight'].values.reshape(-1,1)
x = data['Width'].values.reshape(-1,1)
```

```python
polynomial_regression = PolynomialFeatures(degree = 15)    # 5.mertebeye kadar bakalim. Eger uygun degilse degistirmeliyiz
x_polynomial = polynomial_regression.fit_transform(x)


linear_regression2 = LinearRegression()
linear_regression2.fit(x_polynomial,y)

y_head2 = linear_regression2.predict(x_polynomial)

plt.scatter(x, y, color="blue", label="Actual Data")  # Scatter plot of actual data
plt.plot(x, y_head2, color="green", label="Polynomial Fit")  # Polynomial regression line
plt.xlabel("Width")  # X-axis label
plt.ylabel("Weight")  # Y-axis label
plt.title("Polynomial Regression: Weight vs. Width")  # Title of the plot
plt.legend()
plt.show()
```
